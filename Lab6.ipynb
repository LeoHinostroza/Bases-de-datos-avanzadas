```python
import pandas as pd
from IPython.display import display
import os
```

```python

# cargar el archivo
data = pd.read_csv('sales_data.csv', encoding='ISO-8859-1')  
print("Datos cargados en memoria:")
display(data.head())
```

```python
#Eliminar los datos nulos en state
data = data.dropna(subset=['STATE'])
print("Datos después de eliminar filas con 'state' en null:")
display(data)
```

```python
#crear el archivo xml 
# Ruta de la carpeta de salida
output_folder = './archivo_xml'
os.makedirs(output_folder, exist_ok=True)  

# Ruta completa del archivo XML
output_path = os.path.join(output_folder, 'sales_data_salida.xml')

# Guardar el DataFrame en un archivo XML en la nueva carpeta
data.to_xml(output_path, index=False, parser='etree')
print(f"Archivo XML guardado en: {output_path}")
```
